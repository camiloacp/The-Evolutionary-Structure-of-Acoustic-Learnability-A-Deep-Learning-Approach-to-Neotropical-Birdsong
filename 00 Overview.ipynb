```python
# manejo de datos
import numpy as np
import pandas as pd
import os

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def encontrar_especies_con_pocos_archivos(ruta_base):
    columnas = ['Family', 'Genus', 'Species', 'Comun Name', 'Number of recordings']  # Añadir el nombre de la nueva columna para el nombre común
    datos = []  # Lista para acumular los datos
    
    familias = os.listdir(ruta_base)
    for familia in familias:
        ruta_familia = os.path.join(ruta_base, familia)
        generos = os.listdir(ruta_familia)
        
        for genero in generos:
            ruta_genero = os.path.join(ruta_familia, genero)
            especies = os.listdir(ruta_genero)
            
            for especie in especies:
                ruta_especie = os.path.join(ruta_genero, especie)
                nombres_comunes = os.listdir(ruta_especie)  # Obtener los nombres comunes dentro de la carpeta de la especie
                
                for nombre_comun in nombres_comunes:
                    ruta_nombre_comun = os.path.join(ruta_especie, nombre_comun)
                    archivos = os.listdir(ruta_nombre_comun)
                    
                    if len(archivos) <= 50:
                        # Incluir el nombre común y el conteo de archivos en los datos
                        datos.append([familia, genero, especie, nombre_comun, len(archivos)])
    
    especies_pocas_grabaciones = pd.DataFrame(datos, columns=columnas)
    return especies_pocas_grabaciones

# Uso de la función
ruta_base = 'songs/'
df = encontrar_especies_con_pocos_archivos(ruta_base)
```

```python
print(df.shape)
df = df.sort_values(by='Number of recordings', ascending=True)
df.head(20)
```

```python
df.to_csv('Species_Passeriformes.csv', index=False)
```
