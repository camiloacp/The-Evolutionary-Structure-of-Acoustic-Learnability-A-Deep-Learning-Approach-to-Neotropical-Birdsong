# **Extracción**

```python
family_dict = {
    'Thamnophilidae': thamnophilidae,  
    'Grallaridae': grallaridae,  
    'Furnariidae': furnariidae,  
    'Sapayoidae': sapayoidae,  
    'Tyrannidae': tyrannidae,  
    'Cotingidae': cotingidae,  
    'Pipridae': pipridae,  
    'Tityridae': tityridae,  
    'Vireonidae': vireonidae,  
    'Corvidae': corvidae,  
    'Hirundinidae': hirundinidae,  
    'Troglodytidae': troglodytidae,  
    'Turdidae': turdidae,  
    'Mimidae': mimidae,  
    'Motacillidae': motacillidae,  
    'Thraupidae': thraupidae,  
    'Onychorhynchidae': onychorhynchidae,  
    'Fringillidae': fringillidae,  
    'Estrildidae': estrildidae,  
}
```

```python
ruta = './songs'
```

```python
def extract_data(family_dict):
    for family, genus_list in family_dict.items():
        for genus in genus_list:
            for species in genus:
                genus_name = species.split()[0]
                success = False
                for q_gt in ["D"]:
                    try:
                        q = Query(name=species, q_gt=q_gt)
                        q.retrieve_recordings(multiprocess=True, nproc=10, attempts=10, outdir=f"{ruta}/{family}/{genus_name}/{species}")
                        success = True
                    except Exception as e:
                        print(f"Error con q_gt='{q_gt}' para la especie {species}: {e}")

                if not success:
                    try:
                        q = Query(name=species)
                        q.retrieve_recordings(multiprocess=True, nproc=10, attempts=10, outdir=f"{ruta}/{family}/{genus_name}/{species}")
                    except Exception as e:
                        print(f"No se encontraron cantos para la especie {species}: {e}")
```
