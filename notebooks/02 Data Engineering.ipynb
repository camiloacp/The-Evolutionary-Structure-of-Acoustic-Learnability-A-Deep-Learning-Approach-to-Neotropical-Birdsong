```python
import sys
import os
import pandas as pd
import numpy as np
from scipy import stats

import librosa
import librosa.display

from data_engineering import process_features
```

```python
# Obtener la ruta del directorio raíz
root_dir = os.path.abspath(os.path.join(os.getcwd(), '..')) 
```

```python
familias = sorted([f for f in os.listdir('/Users/camcortes/Documents/birds-sounds/songs/') if os.path.isdir(os.path.join('/Users/camcortes/Documents/birds-sounds/songs/', f))])
for familia in familias:
    generos_path = f'/Users/camcortes/Documents/birds-sounds/songs/{familia}/'
    generos = sorted([g for g in os.listdir(generos_path) if os.path.isdir(os.path.join(generos_path, g)) and not g.startswith('.')])
    for genero in generos:
        especies_path = f'{generos_path}{genero}/'
        especies = sorted([e for e in os.listdir(especies_path) if os.path.isdir(os.path.join(especies_path, e)) and not e.startswith('.')])
        for especie in especies:
            name_comun_path = f'{especies_path}{especie}/'
            name_comuns = sorted([n for n in os.listdir(name_comun_path) if os.path.isdir(os.path.join(name_comun_path, n)) and not n.startswith('.')])
            for name_comun in name_comuns:
                ruta_base = f'{name_comun_path}{name_comun}'
                
                nombre_archivo = f"./Users/camcortes/Documents/birds-sounds/data/{familia}_{genero}_{especie}.csv"
                if not os.path.exists(nombre_archivo):
                    print(f'\t\t\tRuta base: {ruta_base}')
                    try:
                        process_features(
                            f"{ruta_base}/",
                            f"./Users/camcortes/Documents/birds-sounds/data/"
                            f"{familia}_{genero}_{especie}"
                        )
                    except Exception as e:
                        print(f'Error procesando {ruta_base}: {e}')
```

```python
ruta_features = '/Users/camcortes/Documents/birds-sounds/data/'
archivos_csv = os.listdir(ruta_features)
dfs = []
for archivo_csv in archivos_csv:
    # Extraer familia, género y especie del nombre del archivo
    partes_nombre = archivo_csv.split('_')
    familia = partes_nombre[0]
    genero = partes_nombre[1]
    especie = '_'.join(partes_nombre[2:]).replace('.csv', '')  # Asume que el nombre de la especie puede tener guiones bajos

    df = pd.read_csv(os.path.join(ruta_features, archivo_csv))
    # Agregar las columnas con la información extraída
    df['Familia'] = familia
    df['Genero'] = genero
    df['Especie'] = especie

    dfs.append(df)

df_final = pd.concat(dfs, ignore_index=True)
df_final.to_csv('/Users/camcortes/Documents/birds-sounds/notebooks/data/species_features.csv', index=False)
```

```python
# Aplica la lógica usando expresiones de polars
df = df.with_columns(
    pl.when(pl.col('Familia').is_in(tiranni)).then(pl.lit('Tyranni'))
    .otherwise(
        pl.when(pl.col('Familia').is_in(passeri)).then(pl.lit('Passeri'))
        .otherwise(pl.lit(None))
    ).alias('Suborder')
)
```

```python
# Contar el número de registros por suborder
df.group_by('Suborder').agg([
    pl.col('Suborder').count().alias('count')
])
```
