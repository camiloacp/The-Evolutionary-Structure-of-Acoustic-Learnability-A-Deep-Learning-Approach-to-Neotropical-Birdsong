# Explicación detallada de características acústicas en librosa

### Características temporales

- **onset_strength**: Mide la magnitud o intensidad con que comienzan los sonidos en una señal. Detecta los momentos donde hay cambios significativos de energía, como el comienzo de notas o vocalizaciones. En cantos de aves, puede identificar el inicio de cada sílaba o trino.

- **zero_crossing_rate (zcr)**: Cuenta las veces que la señal cruza el valor cero por unidad de tiempo. Valores altos indican sonidos sibilantes o fricativos. En vocalizaciones de aves, los silbidos agudos y limpios tendrán un ZCR diferente a los trinos rápidos o sonidos ásperos.

- **autocorrelation**: Mide la similitud de una señal consigo misma en diferentes desplazamientos temporales. Revela patrones periódicos y es útil para detectar la frecuencia fundamental (tono base) o ritmos repetitivos en los cantos.

### Características tonales/cromáticas

- **chroma_stft**: Proyecta el espectro (usando Short-Time Fourier Transform) en 12 bins que representan las 12 notas musicales, independientemente de la octava. Muestra qué notas están presentes en cada momento.

- **chroma_cqt**: Similar a chroma_stft pero usa Constant-Q Transform, que tiene mejor resolución en frecuencias bajas, importante para analizar tonos graves.

- **chroma_cens**: Versión normalizada de los cromagramas que es más robusta a cambios de timbre y dinámica, útil para comparar patrones tonales.

- **tonnetz**: Transforma los cromagramas en un espacio de 6 dimensiones que representa relaciones armónicas como quintas perfectas, terceras mayores y terceras menores. Captura la estructura armónica subyacente del sonido.

### Características espectrales

- **spectral_centroid**: Representa el "centro de masa" del espectro. Valores altos indican predominancia de frecuencias altas (sonidos brillantes). Puede diferenciar entre cantos agudos y graves.

- **spectral_bandwidth**: Mide la dispersión del espectro alrededor del centroide. Valores altos indican un amplio rango de frecuencias (sonidos ricos o ruidosos).

- **spectral_contrast**: Calcula la diferencia entre picos y valles en el espectro en 7 bandas de frecuencia. Distingue entre sonidos tonales (picos claros) y ruido (espectro plano).

- **spectral_rolloff**: Frecuencia por debajo de la cual se concentra el 85% de la energía espectral. Ayuda a caracterizar la distribución de energía en el espectro.

- **rmse** (Root Mean Square Energy): Mide la energía o volumen de la señal. Útil para detectar secciones silenciosas vs. sonoras o analizar patrones de intensidad en el canto.

### Características basadas en mel
Se basan en la escala mel, que imita cómo el oído humano percibe las frecuencias:

- **mel_spectrogram**: Transforma el espectro a la escala mel (128 bandas), que aproxima mejor la percepción humana de las frecuencias. Las frecuencias se comprimen logarítmicamente, dando más resolución a frecuencias bajas.

- **mfcc** (Mel-Frequency Cepstral Coefficients): 40 coeficientes que representan la forma del tracto vocal (en humanos) o las características tímbricas (en aves). Capturan la "huella digital" acústica de cada especie o individuo.

- **delta_mfcc**: Primera derivada temporal de los MFCCs, captura cómo cambian los coeficientes en el tiempo. Detecta transiciones entre sonidos.

- **delta2_mfcc**: Segunda derivada de los MFCCs, captura la aceleración de los cambios. Útil para caracterizar modulaciones complejas en el canto.

Las características basadas en mel son especialmente valiosas porque aproximan cómo percibimos realmente los sonidos: somos más sensibles a diferencias en frecuencias bajas que en altas. Por ejemplo, distinguimos fácilmente entre 100 Hz y 200 Hz, pero nos cuesta distinguir entre 10,000 Hz y 10,100 Hz. Estas características son fundamentales en reconocimiento de habla, clasificación de sonidos y análisis de cantos de aves.

### Características rítmicas

- **tempogram**: Representación que muestra la prominencia de diferentes ritmos a lo largo del tiempo, con 384 valores que corresponden a diferentes tempos. Permite analizar patrones rítmicos en vocalizaciones repetitivas.

### Estadísticas calculadas para cada característica

Para cada característica se calculan 7 estadísticas:
- **mean**: Valor promedio, indica tendencia central
- **std**: Desviación estándar, mide variabilidad
- **skew**: Asimetría, indica si hay más valores por encima o debajo de la media
- **kurtosis**: Curtosis, mide si hay valores extremos (colas pesadas)
- **median**: Mediana, valor central resistente a outliers
- **min/max**: Valores mínimo y máximo, establecen el rango

Estas características permiten clasificar especies, identificar individuos, analizar comportamientos vocales o detectar patrones acústicos específicos en las vocalizaciones de aves.
