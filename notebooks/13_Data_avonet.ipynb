```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

class OutlierRemoverSimple:
    """
    Versión simplificada y robusta del removedor de outliers
    """

    def __init__(self, data_path=None, df=None):
        if df is not None:
            self.df = df.copy()
        elif data_path is not None:
            self.df = pd.read_csv(data_path)
        else:
            raise ValueError("Debe proporcionar data_path o df")

        self.outliers_removed = {}
        self.cleaned_data = None
        self._clean_initial_data()

    def _clean_initial_data(self):
        """
        Limpia datos iniciales: infinitos, NaN extremos, etc.
        """
        print("🧹 Limpiando datos iniciales...")

        # Obtener columnas numéricas
        numeric_cols = self.get_numeric_columns()

        # Reemplazar infinitos con NaN
        self.df[numeric_cols] = self.df[numeric_cols].replace([np.inf, -np.inf], np.nan)

        # Identificar y limpiar valores extremos (muy grandes)
        for col in numeric_cols:
            if self.df[col].dtype in ['float64', 'float32', 'int64', 'int32']:
                # Calcular límites seguros basados en percentiles
                valid_data = self.df[col].dropna()
                if len(valid_data) > 0:
                    p99 = valid_data.quantile(0.99)
                    p01 = valid_data.quantile(0.01)

                    # Si hay valores extremadamente grandes, cortarlos
                    if p99 > 0 and valid_data.max() > p99 * 1000:
                        self.df.loc[self.df[col] > p99 * 100, col] = np.nan
                    if p01 < 0 and valid_data.min() < p01 * 1000:
                        self.df.loc[self.df[col] < p01 * 100, col] = np.nan

        print(f"  ✅ Datos limpiados. Forma final: {self.df.shape}")

    def get_numeric_columns(self):
        """Obtiene las columnas numéricas excluyendo identificadores"""
        exclude_columns = ['track_index', 'Familia', 'Genero', 'Especie']
        numeric_columns = self.df.select_dtypes(include=[np.number]).columns
        return [col for col in numeric_columns if col not in exclude_columns]

    def select_variables_by_variance(self, n_vars=100):
        """
        Selecciona variables con mayor varianza (más informativas)
        """
        print(f"🔍 Seleccionando {n_vars} variables con mayor varianza...")

        numeric_cols = self.get_numeric_columns()

        # Calcular varianza para cada columna
        variances = {}
        for col in numeric_cols:
            var_val = self.df[col].var()
            if not pd.isna(var_val) and var_val > 0:
                variances[col] = var_val

        # Seleccionar las top N variables por varianza
        sorted_vars = sorted(variances.items(), key=lambda x: x[1], reverse=True)
        selected_vars = [var[0] for var in sorted_vars[:n_vars]]

        print(f"  📊 Seleccionadas {len(selected_vars)} variables")
        return selected_vars

    def remove_outliers_simple_iqr(self, species_column='Especie', k=2.0,
                                  min_outlier_vars=3, max_vars=100):
        """
        Método simple y robusto para eliminar outliers
        """
        print("🔍 Eliminando outliers usando método IQR SIMPLE...")

        # Seleccionar variables importantes
        numeric_cols = self.select_variables_by_variance(n_vars=max_vars)
        print(f"  📊 Usando {len(numeric_cols)} variables")

        cleaned_dfs = []
        species_list = self.df[species_column].unique()

        for species in species_list:
            species_data = self.df[self.df[species_column] == species].copy()
            original_count = len(species_data)

            if original_count < 10:  # Conservar especies con pocos datos
                print(f"  📊 {species}: {original_count} registros (conservando todos - pocos datos)")
                cleaned_dfs.append(species_data)
                self.outliers_removed[species] = {
                    'original': original_count,
                    'cleaned': original_count,
                    'outliers': 0,
                    'percentage_removed': 0.0
                }
                continue

            # Contar outliers por registro
            outlier_counts = pd.Series([0] * len(species_data), index=species_data.index)
            valid_vars = 0

            for col in numeric_cols:
                if col in species_data.columns:
                    col_data = species_data[col].dropna()

                    if len(col_data) >= 5:  # Suficientes datos para calcular IQR
                        Q1 = col_data.quantile(0.25)
                        Q3 = col_data.quantile(0.75)
                        IQR = Q3 - Q1

                        if IQR > 0:  # IQR válido
                            valid_vars += 1
                            lower_bound = Q1 - k * IQR
                            upper_bound = Q3 + k * IQR

                            # Identificar outliers en esta variable
                            col_outliers = (species_data[col] < lower_bound) | (species_data[col] > upper_bound)
                            outlier_counts += col_outliers.fillna(False).astype(int)

            if valid_vars == 0:  # No hay variables válidas
                print(f"  📊 {species}: {original_count} registros (conservando todos - sin variables válidas)")
                cleaned_dfs.append(species_data)
                self.outliers_removed[species] = {
                    'original': original_count,
                    'cleaned': original_count,
                    'outliers': 0,
                    'percentage_removed': 0.0
                }
                continue

            # Determinar umbral adaptativo
            outlier_threshold = min(min_outlier_vars, max(1, valid_vars // 10))

            # Eliminar solo outliers extremos
            extreme_outliers = outlier_counts >= outlier_threshold
            species_cleaned = species_data[~extreme_outliers]

            cleaned_count = len(species_cleaned)
            outliers_count = original_count - cleaned_count

            self.outliers_removed[species] = {
                'original': original_count,
                'cleaned': cleaned_count,
                'outliers': outliers_count,
                'percentage_removed': (outliers_count / original_count) * 100,
                'valid_vars': valid_vars,
                'outlier_threshold': outlier_threshold
            }

            cleaned_dfs.append(species_cleaned)

            print(f"  📊 {species}: {original_count} → {cleaned_count} registros "
                  f"({outliers_count} outliers removidos, {outliers_count/original_count*100:.1f}%) "
                  f"[{valid_vars} vars válidas, umbral: {outlier_threshold}]")

        self.cleaned_data = pd.concat(cleaned_dfs, ignore_index=True)
        return self.cleaned_data

    def remove_outliers_percentile_simple(self, species_column='Especie',
                                         percentile=5, min_outlier_vars=2, max_vars=50):
        """
        Elimina outliers usando percentiles extremos
        """
        print("🔍 Eliminando outliers usando PERCENTILES EXTREMOS...")

        numeric_cols = self.select_variables_by_variance(n_vars=max_vars)
        print(f"  📊 Usando {len(numeric_cols)} variables")

        cleaned_dfs = []
        species_list = self.df[species_column].unique()

        for species in species_list:
            species_data = self.df[self.df[species_column] == species].copy()
            original_count = len(species_data)

            if original_count < 10:
                cleaned_dfs.append(species_data)
                self.outliers_removed[species] = {
                    'original': original_count,
                    'cleaned': original_count,
                    'outliers': 0,
                    'percentage_removed': 0.0
                }
                continue

            outlier_counts = pd.Series([0] * len(species_data), index=species_data.index)
            valid_vars = 0

            for col in numeric_cols:
                if col in species_data.columns:
                    col_data = species_data[col].dropna()

                    if len(col_data) >= 10:
                        valid_vars += 1
                        lower_bound = col_data.quantile(percentile/100)
                        upper_bound = col_data.quantile((100-percentile)/100)

                        col_outliers = (species_data[col] < lower_bound) | (species_data[col] > upper_bound)
                        outlier_counts += col_outliers.fillna(False).astype(int)

            if valid_vars == 0:
                cleaned_dfs.append(species_data)
                self.outliers_removed[species] = {
                    'original': original_count,
                    'cleaned': original_count,
                    'outliers': 0,
                    'percentage_removed': 0.0
                }
                continue

            outlier_threshold = min(min_outlier_vars, max(1, valid_vars // 20))
            extreme_outliers = outlier_counts >= outlier_threshold

            species_cleaned = species_data[~extreme_outliers]
            cleaned_count = len(species_cleaned)
            outliers_count = original_count - cleaned_count

            self.outliers_removed[species] = {
                'original': original_count,
                'cleaned': cleaned_count,
                'outliers': outliers_count,
                'percentage_removed': (outliers_count / original_count) * 100
            }

            cleaned_dfs.append(species_cleaned)

            print(f"  📊 {species}: {original_count} → {cleaned_count} registros "
                  f"({outliers_count} outliers removidos, {outliers_count/original_count*100:.1f}%)")

        self.cleaned_data = pd.concat(cleaned_dfs, ignore_index=True)
        return self.cleaned_data

    def get_summary_statistics(self):
        """Estadísticas resumidas"""
        if not self.outliers_removed:
            print("❌ No se han eliminado outliers aún.")
            return None

        summary_df = pd.DataFrame(self.outliers_removed).T.reset_index()
        base_columns = ['Especie', 'Registros_Originales', 'Registros_Limpios',
                       'Outliers_Removidos', 'Porcentaje_Removido']

        # Ajustar columnas según lo que esté disponible
        summary_df.columns = base_columns[:min(len(base_columns), len(summary_df.columns))] + \
                           list(summary_df.columns[min(len(base_columns), len(summary_df.columns)):])

        print("\n📈 Resumen de eliminación de outliers:")
        print(f"Total de especies procesadas: {len(summary_df)}")
        print(f"Registros originales totales: {summary_df['Registros_Originales'].sum()}")
        print(f"Registros limpios totales: {summary_df['Registros_Limpios'].sum()}")
        print(f"Total de outliers removidos: {summary_df['Outliers_Removidos'].sum()}")
        print(f"Porcentaje promedio removido: {summary_df['Porcentaje_Removido'].mean():.2f}%")
        print(f"Especies con 0% removido: {(summary_df['Porcentaje_Removido'] == 0).sum()}")
        print(f"Especies con 100% removido: {(summary_df['Porcentaje_Removido'] == 100).sum()}")

        return summary_df

    def save_cleaned_data(self, file_path):
        """Guarda los datos limpios"""
        if self.cleaned_data is None:
            print("❌ No hay datos limpios para guardar.")
            return

        self.cleaned_data.to_csv(file_path, index=False)
        print(f"✅ Datos limpios guardados en: {file_path}")
        print(f"   Forma de datos limpios: {self.cleaned_data.shape}")
```

* k = 2.0 (conservador): Detecta ~4% de outliers
* k = 3.0 (muy conservador): Detecta ~1% de outliers

# Explicación detallada de características acústicas en librosa

### Características temporales

- **onset_strength**: Mide la magnitud o intensidad con que comienzan los sonidos en una señal. Detecta los momentos donde hay cambios significativos de energía, como el comienzo de notas o vocalizaciones. En cantos de aves, puede identificar el inicio de cada sílaba o trino.

- **zero_crossing_rate (zcr)**: Cuenta las veces que la señal cruza el valor cero por unidad de tiempo. Valores altos indican sonidos sibilantes o fricativos. En vocalizaciones de aves, los silbidos agudos y limpios tendrán un ZCR diferente a los trinos rápidos o sonidos ásperos.

- **autocorrelation**: Mide la similitud de una señal consigo misma en diferentes desplazamientos temporales. Revela patrones periódicos y es útil para detectar la frecuencia fundamental (tono base) o ritmos repetitivos en los cantos.

### Características tonales/cromáticas

- **chroma_stft**: Proyecta el espectro (usando Short-Time Fourier Transform) en 12 bins que representan las 12 notas musicales, independientemente de la octava. Muestra qué notas están presentes en cada momento.

- **chroma_cqt**: Similar a chroma_stft pero usa Constant-Q Transform, que tiene mejor resolución en frecuencias bajas, importante para analizar tonos graves.

- **chroma_cens**: Versión normalizada de los cromagramas que es más robusta a cambios de timbre y dinámica, útil para comparar patrones tonales.

- **tonnetz**: Transforma los cromagramas en un espacio de 6 dimensiones que representa relaciones armónicas como quintas perfectas, terceras mayores y terceras menores. Captura la estructura armónica subyacente del sonido.

### Características espectrales

- **spectral_centroid**: Representa el "centro de masa" del espectro. Valores altos indican predominancia de frecuencias altas (sonidos brillantes). Puede diferenciar entre cantos agudos y graves.

- **spectral_bandwidth**: Mide la dispersión del espectro alrededor del centroide. Valores altos indican un amplio rango de frecuencias (sonidos ricos o ruidosos).

- **spectral_contrast**: Calcula la diferencia entre picos y valles en el espectro en 7 bandas de frecuencia. Distingue entre sonidos tonales (picos claros) y ruido (espectro plano).

- **spectral_rolloff**: Frecuencia por debajo de la cual se concentra el 85% de la energía espectral. Ayuda a caracterizar la distribución de energía en el espectro.

- **rmse** (Root Mean Square Energy): Mide la energía o volumen de la señal. Útil para detectar secciones silenciosas vs. sonoras o analizar patrones de intensidad en el canto.

### Características basadas en mel
Se basan en la escala mel, que imita cómo el oído humano percibe las frecuencias:

- **mel_spectrogram**: Transforma el espectro a la escala mel (128 bandas), que aproxima mejor la percepción humana de las frecuencias. Las frecuencias se comprimen logarítmicamente, dando más resolución a frecuencias bajas.

- **mfcc** (Mel-Frequency Cepstral Coefficients): 40 coeficientes que representan la forma del tracto vocal (en humanos) o las características tímbricas (en aves). Capturan la "huella digital" acústica de cada especie o individuo.

- **delta_mfcc**: Primera derivada temporal de los MFCCs, captura cómo cambian los coeficientes en el tiempo. Detecta transiciones entre sonidos.

- **delta2_mfcc**: Segunda derivada de los MFCCs, captura la aceleración de los cambios. Útil para caracterizar modulaciones complejas en el canto.

Las características basadas en mel son especialmente valiosas porque aproximan cómo percibimos realmente los sonidos: somos más sensibles a diferencias en frecuencias bajas que en altas. Por ejemplo, distinguimos fácilmente entre 100 Hz y 200 Hz, pero nos cuesta distinguir entre 10,000 Hz y 10,100 Hz. Estas características son fundamentales en reconocimiento de habla, clasificación de sonidos y análisis de cantos de aves.

### Características rítmicas

- **tempogram**: Representación que muestra la prominencia de diferentes ritmos a lo largo del tiempo, con 384 valores que corresponden a diferentes tempos. Permite analizar patrones rítmicos en vocalizaciones repetitivas.

### Estadísticas calculadas para cada característica

Para cada característica se calculan 7 estadísticas:
- **mean**: Valor promedio, indica tendencia central
- **std**: Desviación estándar, mide variabilidad
- **skew**: Asimetría, indica si hay más valores por encima o debajo de la media
- **kurtosis**: Curtosis, mide si hay valores extremos (colas pesadas)
- **median**: Mediana, valor central resistente a outliers
- **min/max**: Valores mínimo y máximo, establecen el rango

Estas características permiten clasificar especies, identificar individuos, analizar comportamientos vocales o detectar patrones acústicos específicos en las vocalizaciones de aves.
