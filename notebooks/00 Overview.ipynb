```python
# manejo de datos
import numpy as np
import pandas as pd
import os

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def encontrar_especies_con_pocos_archivos(ruta_base):
    columnas = ['Family', 'Genus', 'Species', 'Common Name', 'Number of recordings']
    datos = []
    
    familias = [f for f in os.listdir(ruta_base) if os.path.isdir(os.path.join(ruta_base, f))]
    for familia in familias:
        ruta_familia = os.path.join(ruta_base, familia)
        generos = [g for g in os.listdir(ruta_familia) if os.path.isdir(os.path.join(ruta_familia, g))]
        
        for genero in generos:
            ruta_genero = os.path.join(ruta_familia, genero)
            especies = [e for e in os.listdir(ruta_genero) if os.path.isdir(os.path.join(ruta_genero, e))]
            
            for especie in especies:
                ruta_especie = os.path.join(ruta_genero, especie)
                nombres_comunes = [n for n in os.listdir(ruta_especie) if os.path.isdir(os.path.join(ruta_especie, n))]
                
                for nombre_comun in nombres_comunes:
                    ruta_nombre_comun = os.path.join(ruta_especie, nombre_comun)
                    archivos = os.listdir(ruta_nombre_comun)
                    
                    if len(archivos) > 0:
                        datos.append([familia, genero, especie, nombre_comun, len(archivos)])
    
    especies_pocas_grabaciones = pd.DataFrame(datos, columns=columnas)
    return especies_pocas_grabaciones

# Uso de la función
ruta_base = '/Users/camcortes/Documents/birds-sounds/songs'
df = encontrar_especies_con_pocos_archivos(ruta_base)
```

```python
print(df.shape)
df = df.sort_values(by='Number of recordings', ascending=True)
df.head(20)
```
