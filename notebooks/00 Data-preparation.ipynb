```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import taxonomia
import utils

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./data/species_features.csv')
print(df.shape)
df.head()
```

```python
df = df.drop(columns=[
    'track_index'
])
```

```python
df = utils.elimina_nulos(df)
```

```python
species = df['Especie']
df_taxonomy = df[['Especie', 'Genero', 'Familia']]
df = df.drop(columns=['Especie', 'Genero', 'Familia'])
df_taxonomy = df_taxonomy.drop_duplicates()
df_taxonomy.head()
```

```python
df['Specie'] = species
```

```python
df_summary = df.groupby('Specie').mean()
df_summary = df_summary.dropna(axis=1)
df_summary.reset_index(inplace=True)
df_merged = df_summary.merge(df_taxonomy, left_on='Specie', right_on='Especie', how='left')

familia = taxonomia.capitalize_family_genus(taxonomia.genus_family)
df_merged['Family'] = df_merged['Genero'].map(familia)
df_merged['Suborder'] = df_merged['Family'].map(taxonomia.suborder)
df_merged = df_merged.rename(columns={'Genero': 'Genus'})

df_merged.head()
```

```python
df_merged[['Suborder', 'Family', 'Genus']].isnull().sum().sort_values(ascending=False)
```

```python
df_merged['Specie'] = df_merged['Specie'].str.replace(" ", "_")
```

```python
#guarda el csv
df_merged.to_csv('./data/data_summary.csv', index=False)
```
