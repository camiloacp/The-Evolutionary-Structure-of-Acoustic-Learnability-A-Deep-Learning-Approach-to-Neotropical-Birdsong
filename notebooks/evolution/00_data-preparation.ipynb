```python
df = pd.read_csv('/Users/camcortes/Documents/birds-sounds/notebooks/data/species_features.csv')
print(df.shape)
df.head()
```

```python
df = df.drop(columns=[
    'track_index'
])
```

```python
def eliminar_columnas_nulos(df, umbral=0.9):
    porcentaje_nulos = df.isnull().mean()
    columnas_a_eliminar = porcentaje_nulos[porcentaje_nulos > umbral].index
    df = df.drop(columns=columnas_a_eliminar)
    return df
```

```python
species = df.pop('Especie')
genero = df.pop('Genero')
familia = df.pop('Familia')
suborder = df.pop('Suborder')
```
