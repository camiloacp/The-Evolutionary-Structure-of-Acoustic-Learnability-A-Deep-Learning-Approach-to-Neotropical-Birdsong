```python
avonet3 = pd.read_excel("./data_total/AVONET Supplementary dataset 1.xlsx", sheet_name='AVONET3_BirdTree')
```

```python

avonet3 = avonet3[avonet3['Order3'] == 'Passeriformes']
avonet3['Species3'] = avonet3['Species3'].str.replace(" ", "_")
avonet3 = avonet3[['Species3','Species.Status', ]]
print(avonet3.shape)
avonet3.head(3)
```

```python
avonet3 = avonet3.rename(columns={'Species3': 'Species1'})
df_merge, avonet3 = utils.sinonimias(df_merge, avonet3)
```

```python
avonet3 = avonet3.rename(columns={'Species1': 'Specie'})
df_merge = pd.merge(df_merge, avonet3, how="left", on=["Specie", "Specie"])
```

```python
#crea una columna llamada genus haciendo split de la columna Specie
df_merge['Genus'] = df_merge['Specie'].str.split("_").str[0]
```

```python
familia = taxonomia.capitalize_family_genus(taxonomia.genus_family)
df_merge['Family'] = df_merge['Genus'].map(familia)
df_merge['Suborder'] = df_merge['Family'].map(taxonomia.suborder)
```
