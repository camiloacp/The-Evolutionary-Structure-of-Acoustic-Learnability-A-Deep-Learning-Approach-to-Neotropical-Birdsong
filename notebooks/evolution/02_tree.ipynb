```r
# Instalar y cargar el paquete dplyr si no está instalado
if (!require(dplyr)) {
  install.packages("dplyr")
}
library(dplyr)

# Crear un DataFrame de ejemplo
df <- data.frame(
  columna1 = c("A", "B", "C"),
  columna2 = c(1, 2, 3)
)

# Reemplazar el valor "B" en la columna1 por "Z" usando mutate
df <- df %>%
  mutate(columna1 = ifelse(columna1 == "B", "Z", columna1))

print(df)
```
