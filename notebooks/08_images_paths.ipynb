```python
import os
import pandas as pd
```

```python
ruta_base = "./images_spectograms"

def obtener_spectogramas(ruta_base):
    data = []
    
    image_db = os.listdir(ruta_base)
    for i in image_db:
        ruta_db = os.path.join(ruta_base, i)
        for sp in os.listdir(ruta_db):
            ruta_sp = os.path.join(ruta_base, i, sp)
            for img in os.listdir(ruta_sp):
                ruta_img = os.path.join(ruta_base, i, sp, img)
                data.append({
                    "Specie": sp,
                    "image_path": ruta_img,
                    "db": ruta_db.split("/")[-1]
                })
    return pd.DataFrame(data)
```

```python
df = obtener_spectogramas(ruta_base)
```

```python
df[df['db']=='train'].to_csv('./data_spec/train_spectograms.csv', index=False)
df[df['db']=='test'].to_csv('./data_spec/test_spectograms.csv', index=False)
df[df['db']=='val'].to_csv('./data_spec/val_spectograms.csv', index=False)
```
