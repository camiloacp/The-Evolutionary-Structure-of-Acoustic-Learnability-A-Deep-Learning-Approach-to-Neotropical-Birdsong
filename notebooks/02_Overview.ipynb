```python
# manejo de datos
import numpy as np
import pandas as pd
import os

from google.colab import drive
drive.mount('/content/drive')

import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
def summary(ruta_base):
    columnas = ['Family', 'Genus', 'Species', 'Comun Name', 'Number of Recordings']  # Incluir el nombre de la columna corregido
    datos = []  # Lista para acumular los datos

    try:
        familias = os.listdir(ruta_base)
    except Exception as e:
        print(f"Error al listar {ruta_base}: {e}")
        return pd.DataFrame(datos, columns=columnas)

    for familia in familias:
        ruta_familia = os.path.join(ruta_base, familia)
        if not os.path.isdir(ruta_familia):
            continue  # Si no es un directorio, omite

        try:
            generos = os.listdir(ruta_familia)
        except Exception as e:
            print(f"Error al listar {ruta_familia}: {e}")
            continue

        for genero in generos:
            ruta_genero = os.path.join(ruta_familia, genero)
            if not os.path.isdir(ruta_genero):
                continue  # Si no es un directorio, omite

            try:
                especies = os.listdir(ruta_genero)
            except Exception as e:
                print(f"Error al listar {ruta_genero}: {e}")
                continue

            for especie in especies:
                ruta_especie = os.path.join(ruta_genero, especie)
                if not os.path.isdir(ruta_especie):
                    continue  # Si no es un directorio, omite

                try:
                    nombres_comunes = os.listdir(ruta_especie)
                except Exception as e:
                    print(f"Error al listar {ruta_especie}: {e}")
                    continue

                for nombre_comun in nombres_comunes:
                    ruta_nombre_comun = os.path.join(ruta_especie, nombre_comun)
                    if not os.path.isdir(ruta_nombre_comun):
                        continue  # Si no es un directorio, omite

                    try:
                        archivos = os.listdir(ruta_nombre_comun)
                        # Filtrar `.DS_Store` o cualquier otro archivo no deseado
                        archivos = [archivo for archivo in archivos if not archivo.startswith('.')]
                    except Exception as e:
                        print(f"Error al listar {ruta_nombre_comun}: {e}")
                        continue

                    if len(archivos) > 0:
                        # Incluir el nombre común y el conteo de archivos en los datos
                        datos.append([familia, genero, especie, nombre_comun, len(archivos)])

    especies_pocas_grabaciones = pd.DataFrame(datos, columns=columnas)
    return especies_pocas_grabaciones
```

```python
# Uso de la función
ruta_base = '/content/drive/My Drive/songs'
df = summary(ruta_base)
```
