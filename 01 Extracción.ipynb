```python
def extract_data(family_dict):
    for family, genus_list in family_dict.items():
        for genus in genus_list:
            for species in genus:
                genus_name = species.split()[0]
                q = Query(name=species)
                q.retrieve_recordings(multiprocess=True, nproc=10, attempts=10, outdir=f"datasets/{family}/{genus_name}/{species}")
```

```python
extract_data(family_dict)
```
