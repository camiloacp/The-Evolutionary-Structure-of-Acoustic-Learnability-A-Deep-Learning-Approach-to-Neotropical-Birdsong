```python
from datasets import Dataset
import pandas as pd
import os
import librosa

pd.set_option('display.max_columns', None)
```

```python
# Definir la ruta al directorio con los archivos de audio
audio_dir = 'Xiphorhynchus elegans/ElegantWoodcreeper'

# Crear listas para almacenar los datos de cada columna
audios = []
sample_rates = []
paths = []

# Leer los archivos de audio del directorio
for filename in os.listdir(audio_dir):
    if filename.endswith('.mp3'):  # Asegúrate de que es un archivo de audio
        file_path = os.path.join(audio_dir, filename)
        # Cargar el audio con librosa o cualquier otra biblioteca según necesites
        audio, sr = librosa.load(file_path, sr=None)
        # Añadir el audio y cualquier metadato relevante a las listas
        audios.append(audio)
        sample_rates.append(sr)
        paths.append(file_path)

# Crear un diccionario con los datos organizados por columnas
data_dict = {'audio': audios, 'sample_rate': sample_rates, 'path': paths}

# Crear un objeto Dataset
dataset = Dataset.from_dict(data_dict)
# Ahora `dataset` contiene tus audios y está listo para ser procesado o analizado
```

```python
def cargar_audios(ruta_base, especie, genero, familia, audios, sample_rates, paths, species_list, genus_list, familys_list):
	if os.path.isdir(ruta_base):
		for nombre_comun in os.listdir(ruta_base):
			ruta_nombre_comun = os.path.join(ruta_base, nombre_comun)
			if os.path.isdir(ruta_nombre_comun):
				for archivo in os.listdir(ruta_nombre_comun):
					if archivo.endswith('.mp3'):
						ruta_archivo = os.path.join(ruta_nombre_comun, archivo)
						try:
							audio, sr = librosa.load(ruta_archivo, sr=None)
							audios.append(audio)
							sample_rates.append(sr)
							paths.append(ruta_archivo)
							species_list.append(especie)
							genus_list.append(genero)
							familys_list.append(familia)
						except Exception as e:
							print(f"Error al cargar {ruta_archivo}: {e}")
	return audios, sample_rates, paths, species_list, genus_list, familys_list
```

```python
audios = []
sample_rates = []
paths = []
species_list = []
genus_list = []
familys_list = []

familias = os.listdir('cantos_prueba/')
print(f'El numero de familias es: {len(familias)}')
for familia in familias:
    print(f'Procesando familia: {familia}')
    
    # Listar todos los géneros dentro de la familia actual
    generos_path = f'cantos_prueba/{familia}/'
    generos = os.listdir(generos_path)
    print(f'  El número de géneros es: {len(generos)}')
    for genero in generos:
        print(f'  Procesando género: {genero}')
        
        # Listar todas las especies dentro del género actual
        especies_path = f'{generos_path}{genero}/'
        especies = os.listdir(especies_path)
        print(f'    El número de especies es: {len(especies)}')
        for especie in especies:
            print(f'    Procesando especie: {especie}')
            
            # Construir la ruta base para la especie actual
            ruta_base = f'{especies_path}{especie}/'
            print(f'      Ruta base: {ruta_base}')
            audios, sample_rates, paths, species_list, genus_list, familys_list = cargar_audios(ruta_base, especie, genero, familia, audios, sample_rates, paths, species_list, genus_list, familys_list)            
            print(f"Se cargaron {len(audios)} archivos de audio.")
            
            # Verificar si se encontraron audios
            if not audios:
                print(f'No se encontraron cantos para la especie {especie} en {genero}, {familia}')
                continue  # Salta a la próxima especie
```

```python
data_dict = {'audio': audios, 'sample_rate': sample_rates, 'path': paths, 'specie': species_list, 'genus': genus_list, 'family': familys_list}
```
