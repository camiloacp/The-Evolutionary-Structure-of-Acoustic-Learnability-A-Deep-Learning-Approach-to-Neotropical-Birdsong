```python
avonet2 = pd.read_excel("./data_total/AVONET Supplementary dataset 1.xlsx", sheet_name='AVONET2_eBird')
```

```python
avonet2 = avonet2[avonet2['Order2'] == 'Passeriformes']
avonet2['Species2'] = avonet2['Species2'].str.replace(" ", "_")
print(avonet2.shape)
```

```python
col_duplicate = ['Trophic.Level', 'Habitat', 'Habitat.Density', 'Primary.Lifestyle', 'Inference', 'Tail.Length',
                 'Hand-Wing.Index', 'Trophic.Niche', 'Beak.Length_Nares', 'Female', 'Wing.Length', 'Male', 
                 'Mass.Refs.Other', 'Kipps.Distance', 'Unknown', 'Complete.measures', 'Total.individuals', 'Mass', 
                 'Reference.species', 'Migration', 'Tarsus.Length', 'Beak.Length_Culmen', 'Secondary1', 'Beak.Depth', 
                 'Beak.Width', 'Traits.inferred', 'Mass.Source']
```

```python
for col in col_duplicate:
    avonet2 = avonet2.rename(columns={col: col + '_ebird'})
```
