```r
library(tidyverse)
```

```r
data <- read.csv("./data_total/species_features.csv")
```

```r
head(data)
```

```r
# eliminar las columnas familia, genero y suborder
data <- data %>% 
    select(-c(Familia, Genero, Suborder, track_index))
```

```r
head(data, 4)
```

```r
# Agrupar por especie y calcular la media de las variables
data_summary <- data %>%
  group_by(Especie) %>%
  summarise(across(where(is.numeric), \(x) mean(x, na.rm = TRUE)))
```

```r
missing_percentage <- sapply(data_summary, function(x) mean(is.na(x)))

columns_to_keep <- names(missing_percentage[missing_percentage <= 0.8])
data_cleaned <- data_summary[, columns_to_keep]
```

```r
dim(data)
```

```r
dim(data_cleaned)
head(data_cleaned)
```

```r
data_summary[!complete.cases(data_summary),]
```

```r
# filtrar la especie Eremophila alpestris
sp <- data_summary %>% 
    filter(Especie == "Gymnoderus foetidus")
```

```r
# Calcular el número de valores nulos por columna
null_counts <- colSums(is.na(sp))

# Filtrar y mostrar solo las columnas que tienen valores nulos
columns_with_nulls <- null_counts[null_counts > 0]
print(columns_with_nulls)
```

```r
data_final <- data_summary %>% 
    select(-c(names(columns_with_nulls)))
```

```r
species <- data_final$Especie
```

```r
data_final$Especie <- gsub(" ", "_", species)
```
